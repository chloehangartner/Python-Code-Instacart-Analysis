```python
#Count frequency of customers by loyalty flag
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
#Check basis statistics of product prices for each loyalty category
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

```python
#Look at individual customers based on spending to determine is whether the prices of products 
#purchased by loyal customers differ from those purchased by regular or new customers
# First: Create column of mean price by user
ords_prods_merge['mean_order'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.shape
```

```python
# 2: Assign spending flag based on average price

ords_prods_merge.loc[ords_prods_merge['mean_order']<10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_order']>=10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge[['user_id', 'mean_order', 'spending_flag']].head(50)
```

```python
# Create order frequency flag based on median days since previous order
#1: Create a new column showing median days since previous order
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Step two: Assign frequency flag based on median days since previous order
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order']>20, 'Frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order']>10) & (ords_prods_merge['median_days_since_prior_order']<=20), 'Frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order']<=10, 'Frequency_flag'] = 'Frequent customer'
```

```python
#check counts for new customer frequency flag created
ords_prods_merge['Frequency_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'median_days_since_prior_order', 'Frequency_flag']].head(100)
```

```python
# Rename mean_order column to mean_order_price to make it easier to understand
ords_prods_merge.rename(columns = {'mean_order':'mean_order_price'}, inplace = True)
```
